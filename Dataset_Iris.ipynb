#Dataset Iris

```python
import pandas as pd
import numpy as np
```

```python
import sklearn

from sklearn import datasets
```

```python
# importando o dataset iris

iris = datasets.load_iris()
```

```python
# Jogando as informações do dataset iris na variável "dados"

dados = iris.data
```

```python
dados
```

```python
dados.shape
```

```python
# Transformando os dados de iris em DataFrame

dados = pd.DataFrame(dados)
dados
```

```python
iris.target
```

```python
# Jogando as informações do dataset iris.target na variável "classes"

classes = iris.target
classes
```

```python
# Transformando os dados de iris.target em DataFrame chamado "classes"

classes = pd.DataFrame(classes)
classes
```

```python
# Concatenando os 2 DatFrames

df_iris = pd.concat([dados, classes], axis=1)
df_iris
```

```python
# Verificando os nomes originais das colunas

colnames = iris.feature_names
colnames
```

```python
 # Verificando qual a tipagem dos dados presentes em "colnames"

 type(colnames)
```

```python
# Acrescentando uma nova coluna chamada "classification"

colnames.append("Classification")
colnames
```

```python
# Renomeando as colunas conforme os nomes originas

df_iris.rename(columns={0 : "sepal_length"}, inplace=True)
df_iris.rename(columns={1 : "sepal width"}, inplace=True)
df_iris.rename(columns={2 : "petal lenght"}, inplace=True)
df_iris.rename(columns={3 : "petal width"}, inplace=True)
```

```python
# Adicionando os nomes das colunas no DF

df_iris.columns = colnames
df_iris
```

```python
# Verificar o nome conforme as posições / Sendo 0 = 'Setosa', 1 = 'versicolor', 2 = 'virginica'

nomes = iris.target_names
nomes
```

```python
# Substituir as posições pelos nomes

df_iris['Classification'] = df_iris['Classification'].replace({0: 'Setosa', 1: 'versicolor', 2: 'virginica'})
df_iris
```
