```python
import pandas as pd
import numpy as np
```

```python
# importando a tabela

covid_sp = pd.read_csv("/content/drive/MyDrive/Python para dados/dados_covid_sp.csv", sep=";" , encoding="utf8")
```

```python
covid_sp.head(5)
```

```python
# Renomear algumas colunas

covid_sp.rename(columns={"nome_munic" : "municipio", "datahora" : "data", "nome_drs" : "cidade", "cod_drs" : "cod_cidade"}, inplace=True)
```

```python
# Excluindo uma coluna

covid_sp_alterado = covid_sp.drop(columns=["casos_mm7d"])
```

```python
covid_sp_alterado.head(5)
```

```python
# Excluir uma coluna pela posição - Coluna da posição 7

covid_sp = covid_sp.drop(covid_sp.columns[[9]],axis=1)
```

```python
# Formatando os valores da coluna área

covid_sp_alterado["area"] = covid_sp_alterado["area"]/100
```

```python
covid_sp.head(5)
```

```python
# Criação de uma coluna "densidade" (Densidade demográfica)

covid_sp["densidade"] = covid_sp["pop"] / covid_sp["area"]
```

```python
covid_sp.head(5)
```

```python
# Criar uma lista e criar uma coluna de índice a partir dessa lista

lista = list(range(1, 374035))
lista
```

```python
df = pd.DataFrame(lista, columns=["índice"])
df
```

```python
# Unir esse dataframe de índice á tabela de covid

covid_sp = pd.concat([df, covid_sp], axis=1)
covid_sp.head()
```

```python
covid_sp.shape
```

```python
# Localizando e filtrando valores "Ignorado"

covid_sp = covid_sp.loc[covid_sp.municipio != "Ignorado"]
covid_sp
```

```python
# Alterando vírgula por ponto

covid_sp["casos_pc"] = covid_sp["casos_pc"].apply(lambda x: x.replace("," , "."))

covid_sp["casos_mm7d"] = covid_sp["casos_mm7d"].apply(lambda x: x.replace("," , "."))

covid_sp["latitude"] = covid_sp["latitude"].apply(lambda x: x.replace("," , "."))

covid_sp["longitude"] = covid_sp["longitude"].apply(lambda x: x.replace("," , "."))

covid_sp
```

```python
# Excluindo coluna "data"

covid_sp = covid_sp.drop(columns=["data"])
```

```python
# Criando uma coluna com datas , criando DataFrame e concatenando com a tabela covid_sp

data = np.array('2020-02-25', dtype = np.datetime64())
data


data = data + np.arange(579)
data


data = pd.DataFrame(data)
data


data.columns = ["data"]
data.head()


covid_sp = covid_sp.reset_index(drop=True)
covid_sp


covid_sp = pd.concat([data, covid_sp], axis=1)
covid_sp
```
